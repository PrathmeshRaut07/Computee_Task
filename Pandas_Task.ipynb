##Part 2

```python
df.head()
```

```python
top_10_artist=df['artist'].value_counts()
```

```python
top_10_artist=dict(top_10_artist.head(10))
top_10_artist
```

```python
Top_Artist_name=[]
for key,items in top_10_artist.items():
  Top_Artist_name.append(key)
Top_Artist_name

```

```python
Artist=df.groupby(['artist'])
```

```python
Artist['tempo'].mean().loc['Skrillex']
```

```python

Mean=[]
for i in range(len(Top_Artist_name )):
  temp=Artist['tempo'].mean().loc[Top_Artist_name[i]]
  Mean.append(temp)
```

```python
Mean
```

```python
fd=pd.DataFrame({'Top Artist':Top_Artist_name,'Mean of tempo':Mean})
```
